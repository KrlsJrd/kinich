## Numbers




The Python interpreter acts like a simple calculator: you can type an expression into it and it will return the value.

The syntax for expressions is straightforward: the operators `+`, `-`,`*`, `/` and work as in most other languages; parentheses `(())` can also be used for grouping and `**` for exponentiation.

For example:

```python
1+2
```

```python
3-4*5
```

```python
(3-4*5)/2
```

```python
1/2
```

```python
2**2
```

The Integer numbers, as 1,2 y 3 are represented by the type of variable called (`int`), Rational numbers or numbers accompanied with a decimal part, as 5.0 or 1.6 are called type (`float`).



 The division always return a value of type float. e. g.  `1/2` returns `0.5`.



If we want obtain an integer value for division and forget about the rational part, we can use the operator  `//` and for the residue we use the modulo operator `%`.

```python
1//2 # Muestra la parte entera de la division
```

```python
4%3 # Muestra el residuo de la division
```

# Inequalities

Much of programming involves comparing things. The easiest things to compare are numbers.

The standard comparison operators are::

*   `<` (lesser then),
*   `>` (bigger then),
*   `==` (equals to),
*   `<=` (lesser or equal then),
*   `>=` (bigger or equal)
*   `!=` (not equal to).


**Note**: **Do not confuse** `=` with `==`

The first is for assign values to variables.

The second for verify that two things are equal.


e.g.

```python
- 1 < 0
```

```python
5 > (2+1)**2
```

```python
1 == 1
```

```python
3 != 1
```

```python
5 >= 10/2
```

```python
7 <= 11
```

The result of a comparaison always have two possible values:   (`True`) or  (`False`), this variables are called (`Boolean`) type and are the unique of his kind.



The syntax for variables is to write with the first letter capitalized and the rest in lowercase. `True` &`False`.




the booleans could be assign to variables, e.g.:

Another operator that we can use are Otro operador `and` y `or`

for:

*   `and` its mandatory that both comparaison should be `True` to obtain the final result as `True`, if one of the term is not true, then the result will`False`.
*   `or` needs to at least one of the comparaiso  n will be `True` to have the final result as `True`, in other case will be `False`.
